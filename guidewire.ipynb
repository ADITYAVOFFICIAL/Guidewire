```python
# -*- coding: utf-8 -*-
# Install necessary libraries (including kaleido for Optuna plots)
%pip install optuna lightgbm seaborn kaleido -q

import numpy as np
import pandas as pd
import xgboost as xgb
import lightgbm as lgb
import joblib
import os
import time
import optuna # For hyperparameter tuning
import optuna.visualization # Explicit import for visualization functions
import gc # Garbage collection
from sklearn.model_selection import train_test_split # Used for initial split logic reference
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    make_scorer,
    roc_curve,            # For ROC Curve
    precision_recall_curve, # For PR Curve
    auc,                  # For AUC calculation
    ConfusionMatrixDisplay # For Confusion Matrix
)
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import seaborn as sns # For enhanced visualizations
import warnings
import kaleido
# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")


# Suppress specific warnings for cleaner output if desired
# warnings.filterwarnings("ignore", category=FutureWarning)
# warnings.filterwarnings("ignore", category=UserWarning)


print("Libraries imported successfully.")
print(f"XGBoost version: {xgb.__version__}")
print(f"LightGBM version: {lgb.__version__}")
print(f"Optuna version: {optuna.__version__}")

# --- Setup & Configuration ---
print("\n--- Setup & Configuration ---")

# --- !! IMPORTANT: Verify Data Path !! ---
print("Listing directories in /kaggle/input/:")
try:
    print(os.listdir("/kaggle/input/"))
except FileNotFoundError:
    print("Could not list /kaggle/input/. Assuming standard path.")

# --- !! ADJUST this path based on your Kaggle dataset name and structure !! ---
DATA_PATH = "/kaggle/input/alibaba23/cluster-trace-gpu-v2023/csv/"

NODE_FILE = os.path.join(DATA_PATH, "openb_node_list_all_node.csv")
POD_FILE = os.path.join(DATA_PATH, "openb_pod_list_default.csv")

print(f"\nUsing NODE_FILE: {NODE_FILE}")
print(f"Using POD_FILE: {POD_FILE}")
if not os.path.exists(POD_FILE) or not os.path.exists(NODE_FILE):
    print(f"*** ERROR: Data file(s) not found. Please check DATA_PATH. ***")
    raise FileNotFoundError(f"Data file(s) not found at {DATA_PATH}")
# --- End Data Path Verification ---


TARGET_COLUMN = "is_failed"
TIME_COLUMN = "creation_time" # Use creation_time for sorting and splitting

# Model saving prefixes
MODEL_SAVE_PREFIX = "model_"
OPTUNA_STUDY_PREFIX = "optuna_study_"

# Optuna settings
OPTUNA_N_TRIALS_XGB = 25 # Number of Optuna trials for XGBoost
OPTUNA_N_TRIALS_LGB = 25 # Number of Optuna trials for LightGBM
OPTUNA_TIMEOUT = 1500 # Timeout per study (e.g., 25 mins)

# --- GPU Check ---
print("\n--- GPU Check ---")
!nvidia-smi
try:
    # Test XGBoost GPU
    xgb.XGBClassifier(device='cuda')
    XGB_DEVICE = 'cuda'
    print("\nXGBoost GPU check successful ('device=cuda').")
except Exception as e_xgb:
    print(f"\nXGBoost GPU check failed: {e_xgb}. Falling back to CPU.")
    XGB_DEVICE = 'cpu'

try:
    # Test LightGBM GPU
    lgb.LGBMClassifier(device='gpu')
    LGB_DEVICE = 'gpu'
    print("LightGBM GPU check successful ('device=gpu').")
except Exception as e_lgb:
    print(f"\nLightGBM GPU check failed: {e_lgb}. Falling back to CPU.")
    LGB_DEVICE = 'cpu'


# --- Load Data ---
print("\n--- Loading Data ---")
start_time = time.time()
try:
    df_nodes = pd.read_csv(NODE_FILE)
    df_pods_raw = pd.read_csv(POD_FILE)
    print(f"Node data loaded: {df_nodes.shape}")
    print(f"Pod data loaded: {df_pods_raw.shape}")
    print(f"Data loading took: {time.time() - start_time:.2f} seconds")
except Exception as e:
    print(f"Error loading data: {e}")
    raise

# --- Calculate Total Cluster Capacity ---
print("\nCalculating total cluster capacity...")
# Sum capacities, handling potential NaNs in node specs if any
total_cluster_cpu = df_nodes['cpu_milli'].sum()
total_cluster_mem = df_nodes['memory_mib'].sum()
total_cluster_gpu_count = df_nodes['gpu'].sum()
print(f"Total Cluster Capacity: CPU={total_cluster_cpu}m, Mem={total_cluster_mem}MiB, GPUs={total_cluster_gpu_count}")


# --- Preprocessing & Feature Engineering Function ---
def preprocess_and_engineer_features(df_raw, total_cpu, total_mem):
    print("Starting preprocessing and feature engineering...")
    df = df_raw.copy()

    # 1. Target Variable
    df[TARGET_COLUMN] = (df["pod_phase"] == "Failed").astype(int)

    # 2. Timestamps & Basic Durations
    time_cols = [TIME_COLUMN, "deletion_time", "scheduled_time"]
    df[TIME_COLUMN + "_orig"] = df_raw[TIME_COLUMN].copy() # Keep original numeric
    for col in time_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    df['scheduled_time_missing'] = df['scheduled_time'].isna().astype(int)
    df["scheduling_latency_sec"] = df["scheduled_time"] - df[TIME_COLUMN]
    df['scheduling_latency_sec'] = df['scheduling_latency_sec'].fillna(-1)

    # 3. Time-Based Features (Basic + Cyclical)
    df[TIME_COLUMN + "_dt"] = pd.to_datetime(df[TIME_COLUMN], unit='s', errors='coerce')
    df['creation_hour'] = df[TIME_COLUMN + "_dt"].dt.hour.fillna(-1).astype(int)
    df['creation_dayofweek'] = df[TIME_COLUMN + "_dt"].dt.dayofweek.fillna(-1).astype(int)
    # Cyclical encoding
    df['hour_sin'] = np.sin(2 * np.pi * df['creation_hour'] / 24.0)
    df['hour_cos'] = np.cos(2 * np.pi * df['creation_hour'] / 24.0)
    df['dayofweek_sin'] = np.sin(2 * np.pi * df['creation_dayofweek'] / 7.0)
    df['dayofweek_cos'] = np.cos(2 * np.pi * df['creation_dayofweek'] / 7.0)
    df.drop(columns=[TIME_COLUMN + "_dt"], inplace=True) # Drop intermediate datetime

    # 4. Base Numeric Features
    base_numeric_features = ["cpu_milli", "memory_mib", "num_gpu", "gpu_milli"]
    imputer_numeric = SimpleImputer(strategy='median')
    df[base_numeric_features] = imputer_numeric.fit_transform(df[base_numeric_features])

    # 5. Categorical Features
    if df['gpu_spec'].isnull().all():
        df['gpu_spec_provided'] = 0
    else:
        df['gpu_spec_provided'] = (~df['gpu_spec'].isnull()).astype(int)
    df['qos'] = df['qos'].fillna('Unknown')
    df = pd.get_dummies(df, columns=['qos'], prefix='qos', dummy_na=False)

    # 6. Interaction Feature
    df['cpu_x_mem'] = df['cpu_milli'] * df['memory_mib']

    # 7. Cluster State & Advanced Time Features (Requires sorted data)
    print("Calculating cluster state and advanced time features...")
    df.sort_values(by=TIME_COLUMN + "_orig", inplace=True)
    df.reset_index(drop=True, inplace=True) # Ensure RangeIndex after sort
    df[TIME_COLUMN + "_orig"] = pd.to_numeric(df[TIME_COLUMN + "_orig"], errors='coerce')
    df['deletion_time'] = pd.to_numeric(df['deletion_time'], errors='coerce')

    # --- Rolling Failure Rate ---
    # Create DatetimeIndex, handling potential duplicates by keeping original index temporarily
    datetime_index = pd.to_datetime(df[TIME_COLUMN + "_orig"], unit='s', errors='coerce')
    df_temp_for_rolling = df[[TARGET_COLUMN]].copy()
    df_temp_for_rolling.index = datetime_index

    # Check for monotonicity AFTER setting the index
    if df_temp_for_rolling.index.is_monotonic_increasing:
        # Calculate rolling sums/counts on the temporary DataFrame
        rolling_failures = df_temp_for_rolling[TARGET_COLUMN].shift(1).rolling('3600s').sum()
        rolling_count = df_temp_for_rolling[TARGET_COLUMN].shift(1).rolling('3600s').count()
        # Calculate rate, fill NaN, get values (NumPy array)
        failure_rate_values = (rolling_failures / rolling_count).fillna(0).values
        # Assign back using the original DataFrame's index length
        if len(failure_rate_values) == len(df):
             df['recent_failure_rate'] = failure_rate_values
        else:
             print("Warning: Length mismatch during failure rate calculation. Skipping feature.")
             df['recent_failure_rate'] = 0.0
    else:
        print("Warning: Time index not monotonic, skipping rolling failure rate.")
        df['recent_failure_rate'] = 0.0
    del df_temp_for_rolling # Clean up temporary dataframe
    gc.collect()

    # --- Active/Pending Pods & Resource Ratios ---
    active_counts = []
    pending_counts = []
    total_req_cpu = []
    total_req_mem = []
    total_req_gpu_milli = []

    # Ensure time columns are numeric before iteration
    df[TIME_COLUMN + "_orig"] = pd.to_numeric(df[TIME_COLUMN + "_orig"], errors='coerce')
    df['deletion_time'] = pd.to_numeric(df['deletion_time'], errors='coerce')

    for index, row in df.iterrows():
        current_time = row[TIME_COLUMN + "_orig"]
        if pd.isna(current_time):
            active_counts.append(0); pending_counts.append(0); total_req_cpu.append(0); total_req_mem.append(0); total_req_gpu_milli.append(0)
            continue

        # Use .iloc for boolean mask setting to avoid ambiguity with potential duplicate index values during mask creation
        active_mask = (df[TIME_COLUMN + "_orig"] <= current_time) & \
                      (df['deletion_time'].isna() | (df['deletion_time'] > current_time))
        active_mask.iloc[index] = False # Exclude self using positional index

        active_df = df.loc[active_mask] # Use boolean mask with .loc
        active_counts.append(len(active_df))
        total_req_cpu.append(active_df['cpu_milli'].sum())
        total_req_mem.append(active_df['memory_mib'].sum())
        total_req_gpu_milli.append(active_df['gpu_milli'].sum())
        # Count pending among active
        pending_counts.append(active_df['scheduled_time_missing'].sum())

    df['cluster_active_pods'] = active_counts
    df['cluster_pending_pods'] = pending_counts
    df['cluster_req_cpu'] = total_req_cpu
    df['cluster_req_mem'] = total_req_mem
    df['cluster_req_gpu_milli'] = total_req_gpu_milli
    # Calculate Ratios (avoid division by zero)
    df['cluster_cpu_ratio'] = (df['cluster_req_cpu'] / total_cpu).fillna(0) if total_cpu > 0 else 0
    df['cluster_mem_ratio'] = (df['cluster_req_mem'] / total_mem).fillna(0) if total_mem > 0 else 0

    # 8. Define Feature List
    engineered_features = [
        "scheduling_latency_sec", "scheduled_time_missing",
        "creation_hour", "creation_dayofweek", "hour_sin", "hour_cos", "dayofweek_sin", "dayofweek_cos", # Added cyclical
        "gpu_spec_provided",
        "cpu_x_mem",
        "cluster_active_pods", "cluster_pending_pods", # Added pending
        "cluster_req_cpu", "cluster_req_mem", "cluster_req_gpu_milli",
        "cluster_cpu_ratio", "cluster_mem_ratio", # Added ratios
        "recent_failure_rate" # Added failure rate
    ]
    categorical_encoded_features = [col for col in df.columns if col.startswith('qos_')]
    # Ensure base features are correctly listed
    current_base_numeric = ["cpu_milli", "memory_mib", "num_gpu", "gpu_milli"]

    features_to_use = current_base_numeric + engineered_features + categorical_encoded_features
    features_to_use = sorted(list(set(features_to_use)))

    # 9. Final Checks & Cleanup
    nan_counts = df[features_to_use].isnull().sum()
    if nan_counts.sum() > 0:
        print(f"*** WARNING: NaNs found after feature engineering: ***\n{nan_counts[nan_counts > 0]}")
        print("Attempting final imputation...")
        imputer_final = SimpleImputer(strategy='median')
        df[features_to_use] = imputer_final.fit_transform(df[features_to_use])
        if df[features_to_use].isnull().sum().sum() > 0:
             raise ValueError("NaNs still present after final imputation.")
        else:
             print("NaNs filled.")
    else:
        print("No NaNs found in the final feature set.")

    print(f"Feature engineering complete. Number of features: {len(features_to_use)}")
    return df, features_to_use
# --- End of Preprocessing Function ---


# --- Apply Preprocessing ---
start_time_preprocess = time.time()
df_processed, features_to_use = preprocess_and_engineer_features(df_pods_raw, total_cluster_cpu, total_cluster_mem)
print(f"Preprocessing took: {time.time() - start_time_preprocess:.2f} seconds")
print(f"Final processed DataFrame shape: {df_processed.shape}")
gc.collect()


# --- EDA Visualizations ---
print("\n--- Generating EDA Visualizations ---")

# 1. Target Variable Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=TARGET_COLUMN, data=df_processed)
plt.title('Distribution of Target Variable (is_failed)')
plt.xlabel('Pod Failed (1=Yes, 0=No)')
plt.ylabel('Count')
plt.savefig("eda_target_distribution.png")
plt.close() # Close the plot to avoid displaying it inline if not needed
print("Saved eda_target_distribution.png")

# 2. Distribution of Key Numeric Features
key_numeric_features_eda = ['cpu_milli', 'memory_mib', 'scheduling_latency_sec', 'cluster_active_pods', 'recent_failure_rate', 'cpu_x_mem']
plt.figure(figsize=(15, 10))
for i, col in enumerate(key_numeric_features_eda):
    if col in df_processed.columns: # Check if feature exists
        plt.subplot(2, 3, i + 1)
        # Use log scale for highly skewed features if necessary
        if df_processed[col].skew() > 3 or df_processed[col].max() > 1e6: # Example condition for log scale
             plot_data = np.log1p(df_processed[col] - df_processed[col].min() + 1e-6) # Shift potentially negative values
             plot_title = f'Log Distribution of {col}'
        else:
             plot_data = df_processed[col]
             plot_title = f'Distribution of {col}'

        sns.histplot(plot_data, kde=True, bins=50)
        plt.title(plot_title)
        plt.xlabel(col + (' (Log Scale)' if plot_data is not df_processed[col] else ''))
        plt.ylabel('Frequency')
    else:
        print(f"Warning: Feature '{col}' not found for EDA distribution plot.")
plt.tight_layout()
plt.savefig("eda_numeric_distributions.png")
plt.close()
print("Saved eda_numeric_distributions.png")

# 3. Correlation Heatmap
plt.figure(figsize=(12, 10))
# Select only numeric features for correlation calculation
numeric_cols_for_corr = df_processed[features_to_use].select_dtypes(include=np.number).columns
corr = df_processed[numeric_cols_for_corr].corr()
sns.heatmap(corr, cmap='viridis', annot=False, fmt=".2f") # Annot=True can be too crowded
plt.title('Correlation Heatmap of Features')
plt.savefig("eda_correlation_heatmap.png")
plt.close()
print("Saved eda_correlation_heatmap.png")

# 4. Box Plots: Key Features vs. Target
plt.figure(figsize=(15, 10))
for i, col in enumerate(key_numeric_features_eda): # Reuse key features
     if col in df_processed.columns:
        plt.subplot(2, 3, i + 1)
        # Use log scale for features with large ranges/skewness if needed
        log_scale = col in ['memory_mib', 'cpu_milli', 'cpu_x_mem'] and df_processed[col].max() > 10000 # Adjust condition
        temp_df = df_processed.copy()
        # Add small constant to avoid log(0) or log(<0) if using log scale
        if log_scale:
            # Ensure values are positive before log transform
            min_val = temp_df[col].min()
            if min_val <= 0:
                temp_df[col] = temp_df[col] - min_val + 1e-6 # Shift to be positive
            temp_df[col] = np.log1p(temp_df[col])
            plot_title = f'Log({col}) vs. {TARGET_COLUMN}'
            plot_ylabel = f'Log {col}'
        else:
            plot_title = f'{col} vs. {TARGET_COLUMN}'
            plot_ylabel = col

        sns.boxplot(x=TARGET_COLUMN, y=col, data=temp_df, showfliers=False) # Hide outliers for clarity
        plt.title(plot_title)
        plt.xlabel('Pod Failed (1=Yes, 0=No)')
        plt.ylabel(plot_ylabel)
     else:
        print(f"Warning: Feature '{col}' not found for EDA boxplot.")
plt.tight_layout()
plt.savefig("eda_boxplots_feature_vs_target.png")
plt.close()
print("Saved eda_boxplots_feature_vs_target.png")

# --- End EDA ---


# --- Data Splitting (Time-Based) ---
print("\n--- Splitting Data (Time-Based) ---")
start_time_split = time.time()

# Ensure data is sorted by the original numeric time column before splitting
df_processed.sort_values(by=TIME_COLUMN + "_orig", inplace=True)
df_processed.reset_index(drop=True, inplace=True)

X = df_processed[features_to_use]
y = df_processed[TARGET_COLUMN]

# Split: 70% train, 15% validation, 15% test
n_total = len(df_processed)
n_train = int(n_total * 0.70)
n_val = int(n_total * 0.15)
# Ensure test set gets the remainder
n_test = n_total - n_train - n_val

X_train, X_valid, X_test = np.split(X, [n_train, n_train + n_val])
y_train, y_valid, y_test = np.split(y, [n_train, n_train + n_val])

print(f"Train shape: {X_train.shape}, Valid shape: {X_valid.shape}, Test shape: {X_test.shape}")

# --- Scaling Data (for specific models) ---
print("\nScaling data for relevant models...")
# Identify numeric features for scaling (exclude binary/encoded flags)
numeric_features_for_scaling = list(X_train.select_dtypes(include=np.number).columns)
# Remove flags that shouldn't be scaled
flags_to_exclude = ['scheduled_time_missing', 'gpu_spec_provided'] + [c for c in X_train.columns if c.startswith('qos_')]
numeric_features_for_scaling = [f for f in numeric_features_for_scaling if f not in flags_to_exclude]

scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_valid_scaled = X_valid.copy()
X_test_scaled = X_test.copy()

# Fit on train, transform train/valid/test
X_train_scaled[numeric_features_for_scaling] = scaler.fit_transform(X_train[numeric_features_for_scaling])
X_valid_scaled[numeric_features_for_scaling] = scaler.transform(X_valid[numeric_features_for_scaling])
X_test_scaled[numeric_features_for_scaling] = scaler.transform(X_test[numeric_features_for_scaling])
print("Scaling complete.")

print(f"Data splitting & scaling took: {time.time() - start_time_split:.2f} seconds")


# --- Model Training & Evaluation ---
print("\n--- Model Training & Evaluation ---")

results = {} # Dictionary to store results
optuna_studies = {} # Dictionary to store optuna study objects

# --- Function to Evaluate Model & Store Results ---
def evaluate_model(model_name, model, X_tr, y_tr, X_te, y_te, is_scaled=False):
    """Evaluates model on train and test sets and stores results."""
    print(f"\n--- Evaluating {model_name} ---")
    start_time = time.time()
    # Use the correct scaled/unscaled training data for evaluation
    X_train_eval = X_train_scaled if is_scaled else X_tr
    X_test_eval = X_test_scaled if is_scaled else X_te

    test_metrics = {}
    train_metrics = {}
    y_pred_proba_test = None # Initialize

    # Test Set Evaluation
    try:
        y_pred_test = model.predict(X_test_eval)
        y_pred_proba_test = model.predict_proba(X_test_eval)[:, 1]
        test_metrics = {
            'Accuracy': accuracy_score(y_te, y_pred_test),
            'AUC': roc_auc_score(y_te, y_pred_proba_test),
            'Precision': precision_score(y_te, y_pred_test, pos_label=1, zero_division=0),
            'Recall': recall_score(y_te, y_pred_test, pos_label=1, zero_division=0),
            'F1': f1_score(y_te, y_pred_test, pos_label=1, zero_division=0),
            'y_pred_proba': y_pred_proba_test # Store probabilities for plots
        }
        print(f"{model_name} Test Set Performance:")
        print(classification_report(y_te, y_pred_test, target_names=['Not Failed (0)', 'Failed (1)'], zero_division=0))
    except Exception as e:
        print(f"Error during {model_name} test set evaluation: {e}")
        test_metrics = {'Error': f'Test Eval Error: {e}'}


    # Train Set Evaluation (for overfitting check)
    try:
        # Use the correct scaled/unscaled training data for evaluation
        y_pred_train = model.predict(X_train_eval)
        y_pred_proba_train = model.predict_proba(X_train_eval)[:, 1]
        train_metrics = {
            'Accuracy_train': accuracy_score(y_tr, y_pred_train),
            'AUC_train': roc_auc_score(y_tr, y_pred_proba_train),
            'F1_train': f1_score(y_tr, y_pred_train, pos_label=1, zero_division=0)
        }
        print(f"\n{model_name} Train Set Performance:")
        print(f"  Accuracy: {train_metrics['Accuracy_train']:.4f}, AUC: {train_metrics['AUC_train']:.4f}, F1: {train_metrics['F1_train']:.4f}")
    except Exception as e:
        print(f"Error during {model_name} train set evaluation: {e}")
        train_metrics = {'Error': f'Train Eval Error: {e}'}


    # Store results, ensuring no key conflicts if errors occurred
    current_result = results.get(model_name, {})
    current_result.update(test_metrics)
    current_result.update(train_metrics)
    current_result['model'] = model # Always store model if available
    results[model_name] = current_result

    print(f"{model_name} evaluation took: {time.time() - start_time:.2f}s")


# --- Model 1: Logistic Regression (Baseline) ---
print("\n--- Training Logistic Regression ---")
start_time_lr = time.time()
try:
    lr_model = LogisticRegression(solver='liblinear', random_state=42, class_weight='balanced', max_iter=1000)
    lr_model.fit(X_train_scaled, y_train)
    print(f"Logistic Regression trained in {time.time() - start_time_lr:.2f}s")
    # Pass the correct scaled training set (X_train_scaled) for evaluation
    evaluate_model("Logistic Regression", lr_model, X_train_scaled, y_train, X_test_scaled, y_test, is_scaled=True)
except Exception as e:
    print(f"Logistic Regression failed: {e}")
    results['Logistic Regression'] = {'Error': str(e)}


# --- Model 2: Random Forest ---
print("\n--- Training Random Forest ---")
start_time_rf = time.time()
try:
    rf_model = RandomForestClassifier(n_estimators=150, random_state=42, class_weight='balanced',
                                     n_jobs=-1, max_depth=15, min_samples_leaf=5)
    rf_model.fit(X_train, y_train)
    print(f"Random Forest trained in {time.time() - start_time_rf:.2f}s")
    # Pass the original unscaled training set (X_train) for evaluation
    evaluate_model("Random Forest", rf_model, X_train, y_train, X_test, y_test, is_scaled=False)
except Exception as e:
    print(f"Random Forest failed: {e}")
    results['Random Forest'] = {'Error': str(e)}


# --- Shared Optuna Objective Function ---
def create_objective(model_type, X_tr, y_tr, X_va, y_va, scale_pos_w):
    def objective(trial):
        if model_type == 'xgb':
            params = {
                'objective': 'binary:logistic', 'eval_metric': 'auc', 'booster': 'gbtree',
                'device': XGB_DEVICE, 'seed': 42, 'scale_pos_weight': scale_pos_w,
                'lambda': trial.suggest_float('lambda', 1e-7, 10.0, log=True),
                'alpha': trial.suggest_float('alpha', 1e-7, 10.0, log=True),
                'eta': trial.suggest_float('eta', 0.005, 0.1, log=True),
                'max_depth': trial.suggest_int('max_depth', 4, 10),
                'subsample': trial.suggest_float('subsample', 0.4, 0.9),
                'colsample_bytree': trial.suggest_float('colsample_bytree', 0.4, 0.9),
                'min_child_weight': trial.suggest_int('min_child_weight', 1, 30),
                'gamma': trial.suggest_float('gamma', 1e-7, 5.0, log=True)
            }
            model = xgb.XGBClassifier(**params, use_label_encoder=False, early_stopping_rounds=50)
        elif model_type == 'lgb':
            params = {
                'objective': 'binary', 'metric': 'auc', 'boosting_type': 'gbdt',
                'device': LGB_DEVICE, 'seed': 42, 'scale_pos_weight': scale_pos_w, 'n_jobs': -1, 'verbose': -1,
                'lambda_l1': trial.suggest_float('lambda_l1', 1e-7, 10.0, log=True),
                'lambda_l2': trial.suggest_float('lambda_l2', 1e-7, 10.0, log=True),
                'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.1, log=True),
                'num_leaves': trial.suggest_int('num_leaves', 8, 80),
                'max_depth': trial.suggest_int('max_depth', 4, 10),
                'feature_fraction': trial.suggest_float('feature_fraction', 0.4, 0.9),
                'bagging_fraction': trial.suggest_float('bagging_fraction', 0.4, 0.9),
                'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
                'min_child_samples': trial.suggest_int('min_child_samples', 5, 60),
            }
            callbacks = [lgb.early_stopping(stopping_rounds=50, verbose=False)]
            model = lgb.LGBMClassifier(**params)

        try:
            if model_type == 'lgb':
                 model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], callbacks=callbacks)
            else: # xgb
                 model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)

            best_iter = model.best_iteration_ if hasattr(model, 'best_iteration_') and model.best_iteration_ is not None and model.best_iteration_ > 0 else \
                        model.best_iteration if hasattr(model, 'best_iteration') and model.best_iteration > 0 else 1
            trial.set_user_attr('best_iteration', best_iter)

            if model_type == 'lgb':
                 val_score = model.best_score_['valid_0']['auc']
            else: # xgb
                 val_score = model.best_score
            return val_score

        except Exception as e:
            print(f"Trial {trial.number} ({model_type}) failed: {e}")
            trial.set_user_attr('best_iteration', 1)
            return 0.0 # Return low score
    return objective

# --- Calculate scale_pos_weight ---
neg_count_train = (y_train == 0).sum()
pos_count_train = (y_train == 1).sum()
scale_pos_weight_val = neg_count_train / pos_count_train if pos_count_train > 0 else 1
print(f"\nCalculated scale_pos_weight for Boosted Trees: {scale_pos_weight_val:.2f}")

# --- Model 3: XGBoost (Tuned) ---
print("\n--- Tuning and Training XGBoost ---")
start_time_xgb = time.time()
best_params_xgb = {}
best_iteration_xgb = 100 # Default
study_xgb = None # Initialize study object
try:
    study_xgb = optuna.create_study(direction='maximize', study_name=OPTUNA_STUDY_PREFIX + "xgb")
    objective_xgb = create_objective('xgb', X_train, y_train, X_valid, y_valid, scale_pos_weight_val)
    study_xgb.optimize(objective_xgb, n_trials=OPTUNA_N_TRIALS_XGB, timeout=OPTUNA_TIMEOUT)
    optuna_studies['XGBoost'] = study_xgb # Store study object

    if study_xgb.best_trial:
        best_params_xgb = study_xgb.best_params
        best_iteration_xgb = study_xgb.best_trial.user_attrs.get('best_iteration', 100)
        print(f"XGBoost Best Trial AUC: {study_xgb.best_value:.6f}, Best Iteration: {best_iteration_xgb}")
    else:
        print("XGBoost Optuna study finished without successful trials. Using defaults.")

    final_xgb_params = {
        'objective': 'binary:logistic', 'eval_metric': 'auc', 'booster': 'gbtree',
        'device': XGB_DEVICE, 'seed': 42, 'scale_pos_weight': scale_pos_weight_val,
        'n_estimators': best_iteration_xgb, **best_params_xgb
    }
    xgb_model = xgb.XGBClassifier(**final_xgb_params, use_label_encoder=False)
    X_train_valid = pd.concat([X_train, X_valid], axis=0)
    y_train_valid = pd.concat([y_train, y_valid], axis=0)
    xgb_model.fit(X_train_valid, y_train_valid, verbose=False)
    print(f"XGBoost final model trained in {time.time() - start_time_xgb:.2f}s")
    # Pass the original unscaled training set (X_train) for evaluation
    evaluate_model("XGBoost", xgb_model, X_train, y_train, X_test, y_test, is_scaled=False)
    # Store best params found
    if 'XGBoost' in results and isinstance(results['XGBoost'], dict):
        results['XGBoost']['best_params'] = best_params_xgb
        results['XGBoost']['best_iteration'] = best_iteration_xgb

except Exception as e:
    print(f"XGBoost tuning/training failed: {e}")
    results['XGBoost'] = {'Error': str(e)}


# --- Model 4: LightGBM (Tuned) ---
print("\n--- Tuning and Training LightGBM ---")
start_time_lgb = time.time()
best_params_lgb = {}
best_iteration_lgb = 100 # Default
study_lgb = None # Initialize study object
try:
    study_lgb = optuna.create_study(direction='maximize', study_name=OPTUNA_STUDY_PREFIX + "lgb")
    objective_lgb = create_objective('lgb', X_train, y_train, X_valid, y_valid, scale_pos_weight_val)
    study_lgb.optimize(objective_lgb, n_trials=OPTUNA_N_TRIALS_LGB, timeout=OPTUNA_TIMEOUT)
    optuna_studies['LightGBM'] = study_lgb # Store study object

    if study_lgb.best_trial:
        best_params_lgb = study_lgb.best_params
        best_iteration_lgb = study_lgb.best_trial.user_attrs.get('best_iteration', 100)
        print(f"LightGBM Best Trial AUC: {study_lgb.best_value:.6f}, Best Iteration: {best_iteration_lgb}")
    else:
        print("LightGBM Optuna study finished without successful trials. Using defaults.")

    final_lgb_params = {
        'objective': 'binary', 'metric': 'auc', 'boosting_type': 'gbdt',
        'device': LGB_DEVICE, 'seed': 42, 'scale_pos_weight': scale_pos_weight_val, 'n_jobs': -1, 'verbose': -1,
        'n_estimators': best_iteration_lgb, **best_params_lgb
    }
    lgb_model = lgb.LGBMClassifier(**final_lgb_params)
    if 'X_train_valid' not in locals(): # Ensure it exists
         X_train_valid = pd.concat([X_train, X_valid], axis=0)
         y_train_valid = pd.concat([y_train, y_valid], axis=0)
    lgb_model.fit(X_train_valid, y_train_valid, eval_set=[(X_test, y_test)], # Monitor test set during final fit for info
                  callbacks=[lgb.log_evaluation(period=0)]) # Suppress fit logs
    print(f"LightGBM final model trained in {time.time() - start_time_lgb:.2f}s")
    # Pass the original unscaled training set (X_train) for evaluation
    evaluate_model("LightGBM", lgb_model, X_train, y_train, X_test, y_test, is_scaled=False)
    # Store best params found
    if 'LightGBM' in results and isinstance(results['LightGBM'], dict):
        results['LightGBM']['best_params'] = best_params_lgb
        results['LightGBM']['best_iteration'] = best_iteration_lgb

except Exception as e:
    print(f"LightGBM tuning/training failed: {e}")
    results['LightGBM'] = {'Error': str(e)}


# --- Results Comparison ---
print("\n--- Model Comparison ---")
results_df = pd.DataFrame(results).T # Transpose for better readability
# Select relevant metric columns, handle potential errors
metric_cols = ['AUC', 'F1', 'Precision', 'Recall', 'Accuracy', 'AUC_train', 'F1_train'] # Added train metrics
# Filter out rows that only contain errors
valid_results_df = results_df.dropna(subset=[col for col in metric_cols if not col.endswith('_train')], how='all') # Check test metrics
display_cols = [col for col in metric_cols if col in valid_results_df.columns]
if not valid_results_df.empty:
    print(valid_results_df[display_cols].sort_values(by='AUC', ascending=False))
else:
    print("No valid model results to display.")


# --- Performance Visualizations ---
print("\n--- Generating Performance Visualizations ---")

# 1. ROC Curves
plt.figure(figsize=(8, 7))
for name, result in results.items():
    if 'model' in result and 'y_pred_proba' in result and result['y_pred_proba'] is not None:
        fpr, tpr, _ = roc_curve(y_test, result['y_pred_proba'])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.4f})')
    elif 'Error' in result:
         print(f"Skipping ROC for {name} due to error: {result['Error']}")

plt.plot([0, 1], [0, 1], 'k--', label='Chance (AUC = 0.50)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc="lower right")
plt.grid(True)
plt.savefig("performance_roc_curves.png")
plt.close()
print("Saved performance_roc_curves.png")

# 2. Precision-Recall Curves
plt.figure(figsize=(8, 7))
for name, result in results.items():
    if 'model' in result and 'y_pred_proba' in result and result['y_pred_proba'] is not None:
        precision, recall, _ = precision_recall_curve(y_test, result['y_pred_proba'])
        # Calculate AP using sklearn's function if needed, or just plot the curve
        # ap_score = average_precision_score(y_test, result['y_pred_proba'])
        plt.plot(recall, precision, label=f'{name}') # AP score can clutter legend
    elif 'Error' in result:
         print(f"Skipping PR Curve for {name} due to error: {result['Error']}")

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall Curves')
plt.legend(loc="lower left")
plt.grid(True)
plt.savefig("performance_pr_curves.png")
plt.close()
print("Saved performance_pr_curves.png")

# 3. Confusion Matrix (for Best Model based on Test AUC)
best_model_name_cm = None
if not valid_results_df.empty and 'AUC' in valid_results_df.columns:
     if pd.to_numeric(valid_results_df['AUC'], errors='coerce').notna().any():
        best_model_name_cm = valid_results_df['AUC'].astype(float).idxmax()

if best_model_name_cm and best_model_name_cm in results and 'model' in results[best_model_name_cm]:
    print(f"\nGenerating Confusion Matrix for Best Model: {best_model_name_cm}")
    best_model_obj = results[best_model_name_cm]['model']
    is_scaled_best = best_model_name_cm == "Logistic Regression" # Check if scaling needed
    X_test_eval_cm = X_test_scaled if is_scaled_best else X_test
    try:
        y_pred_best = best_model_obj.predict(X_test_eval_cm)
        fig, ax = plt.subplots(figsize=(6, 6))
        ConfusionMatrixDisplay.from_predictions(y_test, y_pred_best, ax=ax, cmap='Blues',
                                                display_labels=['Not Failed', 'Failed'])
        ax.set_title(f'Confusion Matrix - {best_model_name_cm}')
        plt.tight_layout()
        plt.savefig(f"performance_confusion_matrix_{best_model_name_cm.replace(' ', '_')}.png")
        plt.close()
        print(f"Saved performance_confusion_matrix_{best_model_name_cm.replace(' ', '_')}.png")
    except Exception as e:
        print(f"Could not generate confusion matrix for {best_model_name_cm}: {e}")
else:
    print("\nSkipping Confusion Matrix: Best model could not be determined or trained successfully.")


# --- Optuna Visualizations ---
print("\n--- Generating Optuna Visualizations ---")

# 1. Optimization History
for model_name, study in optuna_studies.items():
    if study and study.trials: # Check if study exists and has trials
        try:
            fig = optuna.visualization.plot_optimization_history(study)
            fig.update_layout(title=f'Optuna Optimization History ({model_name})')
            fig.write_image(f"optuna_history_{model_name.replace(' ', '_')}.png")
            print(f"Saved optuna_history_{model_name.replace(' ', '_')}.png")
        except Exception as e:
            print(f"Could not plot Optuna history for {model_name}: {e}")

# 2. Parameter Importance
for model_name, study in optuna_studies.items():
     if study and study.trials:
        try:
            fig = optuna.visualization.plot_param_importances(study)
            fig.update_layout(title=f'Optuna Parameter Importance ({model_name})')
            fig.write_image(f"optuna_param_importance_{model_name.replace(' ', '_')}.png")
            print(f"Saved optuna_param_importance_{model_name.replace(' ', '_')}.png")
        except Exception as e:
            print(f"Could not plot Optuna param importance for {model_name}: {e}")


# --- Feature Importance (Best Model - Already Plotted in Main Section) ---
print("\n--- Feature Importance (Best Model) ---")
# Determine best model based on Test AUC from results_df
if not valid_results_df.empty:
    # Ensure AUC column exists before finding max
    if 'AUC' in valid_results_df.columns:
        # Check if AUC column has valid numeric data before finding idxmax
        if pd.to_numeric(valid_results_df['AUC'], errors='coerce').notna().any():
            best_model_name = valid_results_df['AUC'].astype(float).idxmax()
            print(f"Generating feature importance for best model (based on Test AUC): {best_model_name}")

            if best_model_name in results and 'model' in results[best_model_name]:
                model_to_plot = results[best_model_name]['model']
                try:
                    if hasattr(model_to_plot, 'feature_importances_'):
                        # Use correct feature names from X_train (which has the final columns)
                        feature_names = X_train.columns.tolist()
                        if len(feature_names) == len(model_to_plot.feature_importances_):
                            feature_importance = pd.DataFrame({
                                'feature': feature_names,
                                'importance': model_to_plot.feature_importances_
                            }).sort_values('importance', ascending=False)

                            print(f"Top Features ({best_model_name}):")
                            print(feature_importance.head(20))

                            plt.figure(figsize=(10, 10)) # Adjusted size
                            plot_df = feature_importance[feature_importance['importance'] > 0]
                            if not plot_df.empty:
                                # Use appropriate plotting function based on model type
                                if isinstance(model_to_plot, xgb.XGBClassifier):
                                    xgb.plot_importance(model_to_plot, importance_type='weight', max_num_features=min(30, len(plot_df)), height=0.8, title=f'{best_model_name} Feature Importance (Weight)', show_values=True)
                                elif isinstance(model_to_plot, lgb.LGBMClassifier):
                                    lgb.plot_importance(model_to_plot, max_num_features=min(30, len(plot_df)), height=0.8, title=f'{best_model_name} Feature Importance', importance_type='gain') # Use 'gain' or 'split'
                                elif isinstance(model_to_plot, RandomForestClassifier):
                                     # Simple bar plot for RF
                                     feat_importances = pd.Series(model_to_plot.feature_importances_, index=feature_names) # Use correct index
                                     feat_importances.nlargest(min(30, len(plot_df))).plot(kind='barh', title=f'{best_model_name} Feature Importance')
                                else:
                                     print(f"Feature importance plotting not implemented for {type(model_to_plot)}")

                                plt.tight_layout()
                                plt.savefig(f"{best_model_name}_feature_importance.png")
                                print(f"\n{best_model_name} feature importance plot saved.")
                            else:
                                print("\nNo features with importance > 0 to plot.")
                        else:
                             print(f"Error: Feature name list length ({len(feature_names)}) does not match importance array length ({len(model_to_plot.feature_importances_)}).")
                    else:
                        print(f"{best_model_name} model does not have feature importances attribute.")
                except Exception as e:
                    print(f"Could not generate feature importance for {best_model_name}: {e}")
            else:
                print(f"Could not generate feature importance for {best_model_name} (model not found or failed).")
        else:
            print("Could not determine best model: AUC column contains non-numeric data or is empty.")
    else:
        print("Could not determine best model: AUC column missing from results.")
else:
    print("No valid models trained to determine the best one for feature importance.")


# --- Save Models (Optional) ---
print("\n--- Saving Models (Optional) ---")
for model_name, result_data in results.items():
    # Check if 'model' key exists and the value is not an error string
    if 'model' in result_data and not isinstance(result_data.get('Error', None), str) :
        model_obj = result_data['model']
        try:
            if isinstance(model_obj, xgb.XGBClassifier):
                 filename = f"{MODEL_SAVE_PREFIX}{model_name.replace(' ', '_')}.json"
                 model_obj.save_model(filename)
                 print(f"Saved {model_name} model to {filename}")
            elif isinstance(model_obj, lgb.LGBMClassifier):
                 # Use booster_.save_model for LightGBM's native format
                 filename = f"{MODEL_SAVE_PREFIX}{model_name.replace(' ', '_')}.txt" # Use .txt for native format
                 model_obj.booster_.save_model(filename)
                 print(f"Saved {model_name} model to {filename}")
            else:
                 # Use joblib for other sklearn models
                 filename = f"{MODEL_SAVE_PREFIX}{model_name.replace(' ', '_')}.joblib"
                 joblib.dump(model_obj, filename)
                 print(f"Saved {model_name} model to {filename}")
        except Exception as e:
            print(f"Error saving {model_name} model: {e}")

print("\n--- Script Finished ---")
```
